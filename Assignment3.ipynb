```python
#Question 2)  Write a Python function that checks whether a passed string is palindrome or not
```

```python
def isPalindrome(str):

    for i in range(0, int(len(str)/2)): 
        if str[i] != str[len(str)-i-1]:
            return False
    return True

# main function
s = input("Enter String :")
ans = isPalindrome(s)

if (ans):
    print("String is Palindrome")
else:
    print("String is NotPalindrome")
```

```python
#Question 4) Write a Python function to sum all the numbers in a list
```

```python
def sum_of_list(l,n):
  if n == 0:
    return l[n];
  return l[n] + sum_of_list(l,n-1)

my_list = [1,3,5,2,4,20,34,0,1]
print ("The sum of my_list is :", sum_of_list(my_list,len(my_list)-1))
```

```python
#Question 5 ) Write a Python function to multiply all the numbers in a list
```

```python
def mult_of_list(l,n):
  if n == 0:
    return l[n];
  return l[n] * mult_of_list(l,n-1)

my_list = [1,3,5,2]
print ("The sum of my_list is :", mult_of_list(my_list,len(my_list)-1))
```

```python
#Question 6) Write a Python function that takes a list and returns a new list with unique elements of the first list
```

```python
def unique_list(l):
  num = []
  for a in l:
    if a not in num:
      num.append(a)
  return num
print('unique elements are')
print(unique_list([1,2,4,5,4,5,6,9,10])) 

```
